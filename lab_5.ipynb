```python
import gensim.downloader as api

def solve_analogy_fixed(w1, w2, w3):
    # Використовуємо модель, навчену на Wikipedia (вона набагато точніша)
    print("--- Завантаження якісної моделі (glove-wiki-100)... ---")
    model = api.load("glove-wiki-gigaword-100")

    # Математика: Queen + Man - King = ?
    # Ми шукаємо слово, найближче до цільового вектора
    result = model.most_similar(positive=[w2, w3], negative=[w1], topn=5)

    print(f"\nАналогія: {w1} -> {w2} | {w3} -> ?")
    print("Топ-3 кандидати:")
    for i, (word, similarity) in enumerate(result[:3], 1):
        print(f"{i}. {word} (схожість: {similarity:.4f})")

    return result[0][0]

# Запуск
word = solve_analogy_fixed("doctor", "hospital", "teacher")
```
